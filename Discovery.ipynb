```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/JeffSackmann/tennis_MatchChartingProject/master/charting-m-points-from-2017.csv', encoding= 'unicode_escape')
df.head()
```

```python
# Novak Djokovic vs Andy Murray
nd_am = df[df['match_id'] == '20170107-M-Doha-F-Novak_Djokovic-Andy_Murray']

nd_am = nd_am[['Pt', 'Svr', '1st', '2nd', 'isAce', 'isUnret', 'isRallyWinner', 
               'isForced', 'isUnforced', 'isDouble', 'PtWinner', 'rallyCount']]
nd_am.set_index('Pt', inplace=True)

nd_am['Type_Srv_1st'] = nd_am.iloc[:, 1].str[0]
nd_am['Type_Srv_2st'] = nd_am.iloc[:, 2].str[0]
nd_am['Type_Srv_2st'].fillna(0, inplace=True)

# proximo passo
# devolucao -> se tiver menos de 3 caracteres, o ponto nao seguiu

nd_am
```
