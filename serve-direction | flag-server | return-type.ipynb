# Features

## DireÃ§Ã£o do saque

```python
df_djoko = df[df["match_id"].str.contains("Novak_Djokovic")]

# primeiro caractere do segundo saque
df_djoko['DIR_SRV'] = df_djoko.loc[:, '2nd'].str[0]
# completa com o primeiro caractere do primeiro saque quando ele entrou
df_djoko['DIR_SRV'].fillna(df_djoko.loc[:, '1st'].str[0], inplace=True)

df_djoko.head()
```

## Flag Djoko sacando

```python
df_djoko['DJOKO_SRV'] = df_djoko['Serving'].apply(lambda x: 1 if x == 'ND' else 0)

df_djoko.head()
```

## Devolvedor

```python
df_djoko['RETURN'] = df_djoko.loc[:, '2nd'].str[1:4]
df_djoko['RETURN'].fillna(df_djoko.loc[:, '1st'].str[1:4], inplace=True)

df_djoko.head()
```
