```python
import pandas as pd
```

```python
df = pd.read_csv('raw_data/charting-m-points-from-2017-enriched.csv' , encoding= 'unicode_escape', low_memory=False)
```

```python
df.head()
```

```python
df['Marin Cilic Result'] = pd.Series(pd.NA)
df_player1 = df[df['Player 1'].str.contains("Marin Cilic")]
df_player2 = df[df['Player 2'].str.contains("Marin Cilic")]
```

```python
df_player2.head()
```

```python
df_player1.loc[df_player1['PtWinner'] == 1, 'Marin Cilic Result'] = 1
df_player1.loc[df_player1['PtWinner'] == 2, 'Marin Cilic Result'] = 0
```

```python
df_player2.loc[df_player2['PtWinner'] == 2, 'Marin Cilic Result'] = 1
df_player2.loc[df_player2['PtWinner'] == 1, 'Marin Cilic Result'] = 0
```

```python
df_final = pd.concat([df_player1,df_player2])
```
