```python
#Spliting Serve and Return
df['RETURN'] = df.loc[:, '2nd']
df['RETURN'].fillna(df.loc[:, '1st'], inplace=True)


df['Serve'] = df.loc[:, 'RETURN'].str[0:1]

df['RETURN'] = df['RETURN'].str[1:]

df.head()
```

```python
#Spliting every shot

df['RETURN'] = df['RETURN'].str.replace('f','0f').str.replace('b','0b').str.replace('s','0s').str.replace('r','0r').str.split('0')
df.head(25)
```
