# Dataset

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/JeffSackmann/tennis_MatchChartingProject/master/charting-m-points-from-2017.csv', encoding= 'unicode_escape')
df.head()
```

# Tabela de De-Para

```python
meaning_char = pd.read_csv('https://raw.githubusercontent.com/eduardoscovino/tennis-analytics/master/code_map.csv',
                           encoding= 'unicode_escape', sep=';')
meaning_char
```

# Features

## Direção do saque

```python
df_djoko = df[df["match_id"].str.contains("Novak_Djokovic")]

# primeiro caractere do segundo saque
df_djoko['DIR_SRV'] = df_djoko.loc[:, '2nd'].str[0]
# completa com o primeiro caractere do primeiro saque quando ele entrou
df_djoko['DIR_SRV'].fillna(df_djoko.loc[:, '1st'].str[0], inplace=True)

df_djoko.head()
```

```python
df_djoko = df_djoko.merge(meaning_char[['code', 'info']], left_on='DIR_SRV', right_on='code')
df_djoko['DIR_SRV'] = df_djoko['info']
df_djoko = df_djoko.iloc[:, :-2]
df_djoko
```

## Flag Djoko sacando

```python
df_djoko['DJOKO_SRV'] = df_djoko['Serving'].apply(lambda x: 1 if x == 'ND' else 0)

df_djoko.head()
```

## Devolvedor

```python
df_djoko['RETURN'] = df_djoko.loc[:, '2nd']
df_djoko['RETURN'].fillna(df_djoko.loc[:, '1st'], inplace=True)

df_djoko['Serve'] = df_djoko.loc[:, 'RETURN'].str[0:1]

df_djoko['RETURN'] = df_djoko['RETURN'].str[1:]

df_djoko.head()
```

```python
df_djoko['RETURN'] = df_djoko['RETURN'].str.replace('f','0f').str.replace('b','0b').str.replace('s','0s').str.replace('r','0r').str.replace('@','0@').str.replace('#','0#').str.replace('*','0*').str.replace('n','0n').str.replace('w','0w').str.replace('d','0d').str.replace('x','0x').str.split('0')
```

```python
pd.DataFrame(df_djoko['RETURN'].tolist(), index = df_djoko.index)
```
